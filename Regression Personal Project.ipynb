```python
# Reading in the Excel File
import pandas as pd

df = pd.read_excel (r'C:\Users\student\Desktop\Personal Coding Projects\Regression Python Project\DataSet.xlsx')
print (df)
```

```python
# Here are the 50 lines of different product names
print("Here is a list of individual Sections")
print(df["Product Name"])
print(df["Price"])
print(df["Condition"])
print(df["Seller"])
print(df["Weight"])
print(df["Quantity"])
```

```python
# Counting how many of each Product Name
RiskGames = (df["Product Name"] == 'Risk').sum()
CardGames = (df["Product Name"] == 'Cards').sum()
ScrabbleGames = (df["Product Name"] == 'Scrabble').sum()
MonopolyGames = (df["Product Name"] == 'Monopoly').sum()
ConnectFourGames = (df["Product Name"] == 'ConnectFour').sum()
UnoGames = (df["Product Name"] == 'Uno').sum()
ChessGames = (df["Product Name"] == 'Chess').sum()
CatanGames = (df["Product Name"] == 'Catan').sum()
SorryGames = (df["Product Name"] == 'Sorry').sum()
CheckersGames = (df["Product Name"] == 'Checkers').sum()

print(f'Here are the totals for Products: Risk-{RiskGames} Cards-{CardGames} Scrabble-{ScrabbleGames} Monopoly-{MonopolyGames} Connect Four-{ConnectFourGames} Uno-{UnoGames} Chess-{ChessGames} Catan-{CatanGames} Sorry-{SorryGames} Checkers-{CheckersGames}')

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Product Names')
ax.set_ylabel('Amount of Games')
ax.set_title('Amount of Games by Product Name')
gamesTitle = ['Catan', 'Card', 'Chess', 'Connect Four', 'Monopoly', 'Sorry', 'Scrabble', 'Risk', 'Uno','Checkers']
amountGame = [CatanGames,CardGames,ChessGames,ConnectFourGames,MonopolyGames,SorryGames,ScrabbleGames,RiskGames,UnoGames,CheckersGames]
ax.bar(gamesTitle,amountGame)
plt.show()
```

```python
# Counting how many of price
price1 = (df["Price"] == 10.99).sum()
price2 = (df["Price"] == 11.99).sum()
price3 = (df["Price"] == 12.99).sum()
price4 = (df["Price"] == 13.99).sum()
price5 = (df["Price"] == 14.99).sum()
price6 = (df["Price"] == 15.99).sum()
price7 = (df["Price"] == 16.99).sum()
price8 = (df["Price"] == 17.99).sum()
price9 = (df["Price"] == 18.99).sum()
price10 = (df["Price"] == 19.99).sum()

print(f'Here is a list of the possible prices for Products: {price1},{price2},{price3},{price4},{price5},{price6},{price7},{price8},{price9},{price10}')

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('Prices')
ax.set_ylabel('Frequency of Prices')
ax.set_title('Frequency of Prices by Price')
games = ['10.99', '11.99', '12.99', '13.99', '14.99', '15.99', '16.99', '17.99', '18.99', '19.99']
amounts = [price1,price2,price3,price4,price5,price6,price7,price8,price9,price10]
ax.bar(games,amounts)
plt.show()
```

```python
#Regression
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
amountsArray = np.asarray(amounts)
amountsGameArray = np.asarray(amountGame) 
X = amountsArray.reshape(-1, 1) # values converts it into a numpy array
Y = amountsGameArray.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
Y_pred = linear_regressor.predict(X)  # make predictions

plt.title('Regression Between Frequencies of Amount of Games and Amount of Price')
plt.xlabel('Price Frequencies')
plt.ylabel('Product Name Frequencies')

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.show()

#Conclusion --> Not expecting to see any correlation. All numbers and data points are all random
```
